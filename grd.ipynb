```python
from selenium  import webdriver
```

```python
browser = webdriver.Chrome()
browser.get("https://b.hatena.ne.jp/search/tag?q=COVID-19&sort=recent&users=500&safe=on")
```

```python
art_title_names = browser.find_elements_by_class_name('centerarticle-entry-title')
art_titles = []
for art_title_name in art_title_names:
 value = art_title_name.text
 art_titles.append(value)   
# print(art_titles)
```

```python
bm_user_elements = browser.find_elements_by_class_name('centerarticle-users')
bm_users = []
for bm_user_element in bm_user_elements:
    value = bm_user_element.text.replace('users', '')
    bm_users.append(value)
print(bm_users)
```

```python
post_date_elements = browser.find_elements_by_class_name('entry-contents-date')
post_dates = []
for post_date_element in post_date_elements:
    value = post_date_element.text
    post_dates.append(value)
print(post_dates)
```

```python
get_title_links = browser.find_elements_by_class_name('centerarticle-entry-title')
title_links = []
for title_link in get_title_links:
    get_aTag = title_link.find_element_by_tag_name('a')
    url = get_aTag.get_attribute("href")
    title_links.append(url)
# print(title_links)

```

```python
hoges = browser.find_elements_by_class_name('entrysearch-entry-tags')
for hoge in hoges:
    print(hoge.text)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['タイトル'] = art_titles
df['ブックマーク数'] = bm_users
df['投稿日'] = post_dates
df['詳細リンク'] = title_links
```
